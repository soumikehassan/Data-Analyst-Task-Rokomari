# 2. Price Monitoring Analysis

# Compare the prices of 10 products of Rokomari.com from the dataset with their
# prices on the other 3 e-commerce sites Wafilife, Mamun Book House and
# BoiBazar.

```python
price_columns = df[['ProductName', 'Price_Rokomari', 'Price_Wafilife', 'Price_MamunBookHouse', 'Price_BoiBazar']]

price_comparison_melted = price_columns.melt(id_vars=['ProductName'],
                                             value_vars=['Price_Rokomari', 'Price_Wafilife', 'Price_MamunBookHouse', 'Price_BoiBazar'],
                                             var_name='Ecommerce_Site', value_name='Price')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size and style
plt.figure(figsize=(15, 8))
sns.set(style="whitegrid")
colors = sns.color_palette("viridis")

# Plotting the bar chart using seaborn
sns.barplot(data=price_comparison_melted, x='ProductName', y='Price', hue='Ecommerce_Site', palette=colors)

# Set the title and labels
plt.title('Price Comparison of Top 10 Products Across E-commerce Sites')
plt.xlabel('Product Name')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend(title='E-commerce Site')

# Show the plot
plt.show()
```
